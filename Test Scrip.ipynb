```python
import requests
import json

# Your deployed API URL
API_URL = "https://bosch-pricing-api-2.onrender.com"

def test_prediction():
    print("üöÄ Testing Bosch Pricing API Prediction...")

    # Sample data for prediction
    sample_data = {
        "qty": 25,
        "freight_price": 35.50,
        "product_name_lenght": 45,
        "product_description_lenght": 200,
        "product_photos_qty": 3,
        "product_weight_g": 1500,
        "product_score": 4.2,
        "customers": 5000,
        "weekday": 3,
        "weekend": 0,
        "holiday": 0,
        "month": 11,
        "year": 2024,
        "s": 0.75,
        "volume": 2500,
        "comp_1": 120.0,
        "ps1": 4.0,
        "fp1": 25.0,
        "comp_2": 115.0,
        "ps2": 4.2,
        "fp2": 22.0,
        "comp_3": 125.0,
        "ps3": 3.8,
        "fp3": 28.0,
        "lag_price": 118.0,
        "product_id_encoded": 15,
        "product_category_encoded": 8,
        "price_to_comp1_ratio": 1.0,
        "price_to_comp2_ratio": 1.05,
        "price_to_comp3_ratio": 0.95,
        "avg_competitor_price": 120.0,
        "price_vs_avg_comp": 1.0,
        "total_freight": 45.0,
        "avg_competitor_score": 4.0,
        "price_per_gram": 0.03,
        "demand_indicator": 0.8
    }

    try:
        # Make prediction request
        response = requests.post(f"{API_URL}/predict", json=sample_data)

        if response.status_code == 200:
            result = response.json()
            print("‚úÖ PREDICTION SUCCESS!")
            print(f"üìä Predicted Price: ${result['predicted_price']:.2f}")
            print(f"‚úÖ Status: {result['status']}")
            print(f"üîß Features Used: {result['features_used']}")
        else:
            print(f"‚ùå Prediction failed with status: {response.status_code}")
            print(f"Error: {response.text}")

    except Exception as e:
        print(f"‚ùå Request failed: {e}")

def quick_test():
    """Simpler test with minimal data"""
    print("\nüîç Quick Test with Minimal Data...")

    minimal_data = {
        "qty": 10,
        "freight_price": 25.0,
        "product_score": 4.0,
        "customers": 3000
    }

    try:
        response = requests.post(f"{API_URL}/predict", json=minimal_data)
        print("Quick Test Result:", response.json())
    except Exception as e:
        print(f"Quick test failed: {e}")

if __name__ == "__main__":
    test_prediction()
    quick_test()
```
